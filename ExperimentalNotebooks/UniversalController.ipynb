```python
import torch
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
P = torch.Tensor([2.0, 1.0])
O = torch.Tensor([0.0, 0.0])
plt.plot([O[0], P[0]], [O[1], P[1]], marker='o')
plt.text(float(P[0]) + 0.05, float(P[1]) + 0.05, 'P')
plt.text(float(O[0]) + 0.05, float(O[1]) + 0.05, 'O')
plt.show()
```
